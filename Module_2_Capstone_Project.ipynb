## 3. ***Data Wrangling***

### Data Wrangling Code

````python
df = merged_df.copy()

print(df.shape)
````

````python
round((df.isnull().sum()/len(df))*100,2)
````

````python
df['description'] = df['description'].fillna('Not Available')  # Filling missing description values , because the missing value in this column was negligible.
````

````python
df['age_certification'] = df['age_certification'].fillna('Unknown') # Filling missing age certification values , beacuse missing value in this column was negligible.
````

````python
df[['imdb_score','imdb_votes','tmdb_score','tmdb_popularity']].describe()
````

````python
# Filling numerical missing values with median

df['imdb_score'] = df['imdb_score'].fillna(df['imdb_score'].median())

df['imdb_votes'] = df['imdb_votes'].fillna(df['imdb_votes'].median())

df['tmdb_score'] = df['tmdb_score'].fillna(df['tmdb_score'].median())

df['tmdb_popularity'] = df['tmdb_popularity'].fillna(df['tmdb_popularity'].median())
````

````python
df['name'] = df['name'].fillna('Unknown')

df['role'] = df['role'].fillna('Unknown')

df['person_id'] = df['person_id'].fillna(0)

df['character'] = df['character'].fillna('Not Applicable')
````

````python
df.shape
````

### What all manipulations have you done and insights you found?

Created a copy of the merged dataset for data cleaning and analysis. Checked duplicate records and found that duplicates were generated due to the one-to-many relationship between titles and cast/crew members after merging the datasets. Therefore, duplicates were retained as valid observations. Missing values were analyzed and handled appropriately. The description column was filled with 'Not Available', age_certification with 'Unknown', character with 'Not Applicable', and missing values in person-related columns were replaced with suitable placeholders. Missing values in seasons were retained because movies do not have seasons, and missing values in imdb_id were retained as they do not significantly affect the analysis. After data cleaning, almost all important analytical columns became complete and ready for exploratory data analysis.
